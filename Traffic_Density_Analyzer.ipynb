```python
# 🚀 Step 3.2.2.1 Clone YOLOv5 repo
!git clone https://github.com/ultralytics/yolov5
%cd yolov5

# 🔧 Step 3.2.2.2 Install requirements
!pip install -r requirements.txt

# ✅ Step 3.2.2.3 Import libraries
import torch
import cv2
from google.colab.patches import cv2_imshow

# 🔍 Step 3.2.2.4 Check GPU availability (for faster inference)
print("CUDA available:", torch.cuda.is_available())

# 🔭 Step 3.2.2.5 Load YOLOv5 pretrained model
# Why yolov5s? Small, fast, and detects COCO classes including vehicles
model = torch.hub.load('ultralytics/yolov5', 'yolov5s')

# 📂 Step 3.2.2.6 Upload your downloaded video file from Dataset 1
from google.colab import files
uploaded = files.upload()

video_path = list(uploaded.keys())[0]
print(f"Processing video: {video_path}")

# 🎥 Step 3.2.2.7 Initialize video capture
cap = cv2.VideoCapture(video_path)

frame_count = 0
vehicle_counts = []

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # 🔎 Step 3.2.2.8 YOLOv5 inference per frame
    results = model(frame)
    df = results.pandas().xyxy[0]

    # 🛻 Step 3.2.2.9 Filter for vehicle classes only
    vehicle_df = df[df['name'].isin(['car', 'bus', 'truck', 'motorcycle', 'bicycle'])]
    vehicle_count = len(vehicle_df)
    vehicle_counts.append(vehicle_count)

    # 🖼️ Step 3.2.2.10 Display frame every ~30 frames (avoid overload)
    if frame_count % 30 == 0:
        annotated_frame = results.render()[0]
        cv2_imshow(annotated_frame)
        print(f"Frame {frame_count} | Vehicle count: {vehicle_count}")

    frame_count += 1

cap.release()

# 📊 Step 3.2.2.11 Summary statistics
print(f"\nProcessed {frame_count} frames.")
print(f"Average vehicles per frame: {sum(vehicle_counts) / len(vehicle_counts):.2f}")
```

```python
# 🚀 Step 4.1.1 Import plotting library
import matplotlib.pyplot as plt

# 🔍 Step 4.1.2 Create x-axis frame indices
frames = list(range(len(vehicle_counts)))

# 📊 Step 4.1.3 Plot vehicle counts per frame
plt.figure(figsize=(12,6))
plt.plot(frames, vehicle_counts, color='royalblue', linewidth=2)

# ✨ Step 4.1.4 Label the plot
plt.title("Traffic Density over Video Frames", fontsize=16)
plt.xlabel("Frame Number", fontsize=14)
plt.ylabel("Number of Vehicles Detected", fontsize=14)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()

# 🖼️ Step 4.1.5 Show the plot
plt.show()
```
